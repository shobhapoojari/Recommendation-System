```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
book_reader = book_clean.pivot(index='User',columns='Book',values='Rating').reset_index(drop=True)
book_reader
```

```python
book_reader.fillna(0,inplace=True)
book_reader
```

```python
book1 = 1-pairwise_distances(book_reader.values,metric='cosine')
book1
```

```python
book1_df=pd.DataFrame(book1)
```
